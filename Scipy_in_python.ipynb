```python
# OPtimizers in SCipy: they are a set of procedures to solve the copmplex problem.OPtimizer is like a function
#For Example:
# Root of an equation: x + cos(x), we will solve it via optimize.root function.
# this function takes 2 arguments: "fun" and x0
# Here we will find the root of the equation x + cos(x) :
from scipy.optimize import root
from math import cos
```

```python
def eqn(x):
    return x + cos(x)
myroot = root(eqn, 0)
print(myroot.x)
```

```python
# Minimizing the Function or data:
# for finding the Minima:
# we use scipy.optimize.minimize().
#it uses 3 arguments: "fun", x0 and method:it also has some legal values: "CG", "BFGS", "NEWTON-CG", "L-BFGS-B", "TNC", "COBYLA", "SLSQP".
# Example of the above: in which we will minimize the function: x^2 + x + 2 with BFGS:
from scipy.optimize import minimize
def eqn(x):
    return x**2 + x + 2
mymin = minimize(eqn, 0, method='BFGS')
print(mymin)
```

```python
# What is sparse data: Sparse data is the data that  do not contain  any of the information. like the elements that dont carry any info. [1,0,2,0,4,5,6,9,2,4]
# sparse data example [1,0,2,0,0,3,0,0,0,0,0,4]
# Dense Array: [2,5,6,8,9,7,1,2,3,6]
# How to work with Sparse Data.
# scipy has a module scipy.sparse function. there are 2 matrics in this sparse: CSC(Compressed sparse Column) and CSR(Commpresses Sparse Row)# CSR Matrics: here we will create a CSR matrics from an array:
# CSR Matrics: here we will create a CSR matrics from an array:
import numpy as np
from scipy.sparse import csr_matrix
arr= np.array([0,0,0,0,0,1,1,0,2])
print(csr_matrix(arr))
```

**Scipy Graph:**

```python
# working with graphs: we have a module named scipy.sparse.csgraph
# adjacency Matrix: describe relationships between vertices (nodes) in a graph.nxn matrix where n is the number of element in the graph.
# Connected Component:
from scipy.sparse.csgraph import connected_components
#connected_component is converted in to the matrix so we need to import csr_matrix
from scipy.sparse import csr_matrix
array = np.array([[0,1,2], [1,0,0], [2,0,0]])
new_array = csr_matrix(array)
print(connected_components(new_array))
# single connected component in the graph, and all nodes belong to that component (component label 0).
```

```python
# Dijkstra method: for finding the shortest path in a graph from one element(node) to another
# it take 3 arg: return_predecessors, indices, limit
# here we will find te shortest path from element 1 to 2:
from scipy.sparse.csgraph import dijkstra
#dijkstra function computes the shortest paths from the source node (node 0) to all other nodes in the graph
print(dijkstra(new_array, return_predecessors=True, indices=0))
# returns two arrays: the first array contains the distances of the shortest paths, and the second array contains the predecessor (parent) nodes for each node on the shortest paths.
```

```python
# Floyd Warshall() method: it is for finding the shortest path between all the pairs of elements.
from scipy.sparse.csgraph import floyd_warshall
print(floyd_warshall(new_array, return_predecessors=True))
#return a tuple containing two arrays: the first array contains the shortest distances between all pairs of nodes,
#the second array contains the predecessor nodes for each pair of nodes on the shortest paths.
```

```python
# Bellman_ford() method: it is for finding the shortest path between all the pairs of elements but this method can handle negative weight as well:
from scipy.sparse.csgraph import bellman_ford
array = np.array([[0,-1,2], [1,0,0], [2,0,0]])
new_array = csr_matrix(array)
print(bellman_ford(new_array, return_predecessors=True, indices=0))
#Bellman-Ford()returns two arrays: the first array contains the distances of the shortest paths, and the second array contains the predecessor (parent) nodes for each node on the shortest paths.
```

```python
# Depth first order: it returns a depth first traversal froma node: it has 2 argu: the graph, starting element
from scipy.sparse.csgraph import depth_first_order
arr = np.array([[0,1,0,1], [1,1,1,1], [2,1,1,0],[0,1,0,1]])
new_arr= csr_matrix(arr)
print(depth_first_order(new_arr, 1))
#depth_first_order() returns a tuple containing two arrays: the first array is the list of nodes visited during the depth-first traversal,
#and the second array is the order in which the nodes were traversed.
```

```python
# breadth_first_order() method: it returns the breadth from a node:it has 2 argu: the graph, starting element
from scipy.sparse.csgraph import breadth_first_order
print(breadth_first_order(new_arr, 1))
#This breadth_first_order() returns a tuple containing two arrays: the first array is the list of nodes visited during the breadth-first traversal,
#the second array is the order in which the nodes were traversed.
```

```python
# Working with spatial data: describes the physical location and shape of objects it refers to data that is represented in a geometric space.
#the space b/w the points are called spatial data.
# Triangulation: one method to generate these triangulation through the point is delaunay() triangulation.

# here now we will create a triangualtion from some points:
from scipy.spatial import Delaunay
import matplotlib.pyplot as plt

array = np.array([[2,4], [3,4], [3,0], [2,2], [4,1],[9,7]])
simplices = Delaunay(array).simplices   #simplices is used to break down into differnt parts.
plt.triplot(array[:,0], array[:,1], simplices)
#The triplot function is used to plot the triangles formed by the simplices indices.
plt.scatter(array[:,0], array[:,1], color='r')
plt.show()
```

```python
# Convex HUll: it is the smallest polygon that covers all of the given point:
from scipy.spatial import ConvexHull
import matplotlib.pyplot as plt

array = np.array([[2,4], [3,4], [3,0], [2,2], [4,1], [1,2], [5,0], [3,1],[1,2],[0,2],[5,12],[9,7]])

hull = ConvexHull(array)
hull_points = hull.simplices
plt.scatter(array[:,0], array[:,1])
for simplex in hull_points:      # connect the pints
    plt.plot(array[simplex,0], array[simplex,1], 'k-') #'k-' specifies a black line.
plt.show()
```

```python
# KDTrees: they are a data structures optimized for the nearest neighbour queries.
from scipy.spatial import KDTree
array = [(1,-1), (2,3), (-2,3), (2,-3),(7,2)]
tree = KDTree(array)
res = tree.query((7,2)) # query  find the distance.
#query((1,1)) find the K_nearest nieghbour of that point
print(res)
# two elements: the distance to the nearest neighbor and the index of the nearest neighbor point in the original array list.
```

```python
# Distance matrix: it is used to find the various types of distance between 2 points . there are basically 2 types: Euclidean Distance, Cosine Distance.
# Euclidean Distance:
from scipy.spatial.distance import euclidean
p1 = (1,0)
p2 = (10,2)
dis = euclidean(p1, p2)
dis
```

```python
# Cosine Distance:
from scipy.spatial.distance import cosine
p1 = (1,0)
p2 = (10,2)
dis = cosine(p1, p2)
print(dis)
```

#Scipy Interpolation:

```python
# What is interpolation: it is a method for generating points between the given points. and the method of filling the values is called imputation.
# 1D interpolation: interp1d()
# here for given xs and ys interpolate values from 2.1, 2.2, 2.3........2.9:
from scipy.interpolate import interp1d
xs = np.arange(10)
ys = 2*xs + 1
interp_func = interp1d(xs, ys)
array = interp_func(np.arange(2.1, 3, 0.1))
print(array)
```

```python
# spline interpolation:
# example: here now we will find the univariate spline interpolation for 2.1, 2.2, 2.3.....2.9 .

from scipy.interpolate import UnivariateSpline
xs = np.arange(10)
ys = xs**2 + np.sin(xs) + 1

interp_func = UnivariateSpline(xs,ys)
array= interp_func(np.arange(2.1, 3, 0.1))

print(array)
```

```python
# Interpolation with Radial basis Function(RBF)

from scipy.interpolate import Rbf
xs = np.arange(10)
ys = xs**2 + np.sin(xs) + 1

interp_func = Rbf(xs,ys)
array = interp_func(np.arange(2.1, 3, 0.1))

print(array)
```
